```python
import pandas as pd
from pathlib import Path

cols = [
    "age","workclass","fnlwgt","education","education_num","marital_status",
    "occupation","relationship","race","sex","capital_gain","capital_loss",
    "hours_per_week","native_country","income"
]

train_path = Path("data/adult.data")
test_path  = Path("data/adult.test")

train_df = pd.read_csv(train_path, header=None, names=cols, skipinitialspace=True)
test_df  = pd.read_csv(test_path, header=None, names=cols, skiprows=1, skipinitialspace=True)

# Clean labels/spaces
train_df["income"] = train_df["income"].str.strip()
test_df["income"]  = test_df["income"].str.replace(".", "", regex=False).str.strip()

print("train:", train_df.shape, "test:", test_df.shape)
train_df.head()
```
